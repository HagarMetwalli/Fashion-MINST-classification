```python
class net(nn.Module):
    def __init__():
        super(net,self).__init__()
        self.conv1 = nn.Conv2d(1,16,3,padding=1) # shape = 16 * 28 *28
        self.conv2 = nn.Conv2d(16,32,3,padding=1) # shape = 32 * 28 *28
        self.pool = nn.MaxPool2d(2,2)  # shape (32, 14, 14)
        self.conv3 = nn.Conv2d(32,64,3) # shape (64, 12, 12)
        self.fc = nn.Linear(64*12*12,10)
    
    def forward(self,x):
        x = self.conv1(x)
        x = F.relu(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x= F.relu(self.conv3(x))
        x = x.view(x.size(0),-1)
        x = F.softmax(self.fc(x),dim=1)
        
        
```
